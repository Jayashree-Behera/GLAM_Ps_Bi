```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Bk5=np.loadtxt("Bkavg_z0.50.DAT")
Bk73=np.loadtxt("Bkavg_z0.73.DAT")
```

```python
ratio5=Bk5[:,3]/Bk5[:,4]
var5=Bk5[:,5]
ratio73=Bk73[:,3]/Bk73[:,4]
var73=Bk73[:,5]
```

```python
plt.axhline(1.0, ls=':')
plt.plot(Bk5[:,0],ratio5,".",label="z=0.5")
#plt.ylabel("ratio/error")
plt.ylim(0.8,1.2)
plt.legend()
#plt.savefig("ratio_z0.50.png")
```

```python
plt.hist(np.log(var73),bins = 40,label="z0.73")
plt.hist(np.log(var5),bins = 40,alpha = 0.5,label="z0.5")
plt.legend()
plt.savefig("var_hist.png")
```

```python
arr = np.where((np.log(var5) < -2.1) & (np.log(var5) > -2.25) )
```

```python
arr73 = np.where((np.log(var73) < -2.25) & (np.log(var73) > -2.42) )
```

```python
plt.figure(figsize = (20,10))
v5 = var5/max(var5)
v73 = var73/max(var73)
plt.scatter(x = Bk5[:,0], y = ratio5/0.97,c = np.abs(np.log(v5)),label = "z0.5", s = 60, cmap='Greens')
plt.colorbar()
plt.scatter(x = Bk5[:,0], y = ratio73,c = np.abs(np.log(v73)),label = "z0.73", s = 10, cmap='Reds')
plt.ylim(0.8,1.2)
plt.colorbar()
plt.xlabel("k1")
plt.ylabel("ratio")
#plt.xlim(400,2000)
plt.legend()
plt.savefig("ratio_color.png")
```

```python
plt.figure(figsize = (15,10))
plt.scatter(x=range(len(ratio5)),y = ratio5/(np.abs(np.log(v5))+1e-14),c = np.log(v5),label = "z0.5", s = 30, cmap='Greens_r')
plt.colorbar()
plt.scatter(x=range(len(ratio5)), y = ratio73/(np.abs(np.log(v73))+1e-14),c = np.log(v73),label = "z0.73", s = 10, cmap='Reds_r')
plt.ylim(0.03,0.2)
plt.colorbar()
plt.xlabel("index")
plt.ylabel("ratio/abs(log(var))")
# plt.xlim(400,2000)
plt.legend()
plt.savefig("ratio_by_var.png")
```
